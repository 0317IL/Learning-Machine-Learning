## Importação

```python
import matplotlib.pyplot as plt
```

## Ferramentas

### Plot()
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot

```python
x = [1, 2, 3, 4]
y = [10, 3, 2, 8]

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x, y, label='dados')
plt.ylabel('Eixo Y')
plt.xlabel('Eixo X')
plt.title('Título')
plt.legend()
plt.show()
```

### Lines propriedades
https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html#matplotlib.lines.Line2D

```python
plt.subplot(2, 2, 1)
plt.plot(x, y, linestyle='dashed', color='r')
plt.title("Dashed")

plt.subplot(2, 2, 2)
plt.plot(x, y, linestyle=':', color='g', lw=3.0)
plt.title("Dotted")

plt.subplot(2, 2, 3)
plt.plot(x, y, linestyle='-.', color='y', lw=0.5)
plt.title("Dashdot")

plt.subplot(2, 2, 4)
plt.plot(x, y, linestyle='none')
plt.title("None")
plt.tight_layout()
plt.show()
```

### .Scatter()
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html?highlight=scatter#matplotlib.pyplot.scatter

```python
plt.scatter(x, y, color='r')
plt.show()
```

### .Bar
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html?highlight=bar#matplotlib.pyplot.bar

```python
plt.bar(x, y, color='k')
plt.show()
```

## Nomeando os eixos do gráfico:

### Mudando os tipos de marcadores
https://matplotlib.org/stable/api/markers_api.html?highlight=markers#module-matplotlib.markers

```python
plt.scatter(x, y, color='r', marker='x')
plt.show()
```

### Alterando os limites 
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.axis.html

```python
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.axis(xmin=-1, xmax=10, ymin=0, ymax=12)

plt.subplot(2, 2, 2)
plt.plot(x, y)
plt.axis('square')


plt.subplot(2, 2, 3)
plt.plot(x, y)
plt.axis('image')


plt.subplot(2, 2, 4)
plt.plot(x, y)
plt.axis('tight')
plt.show()
```

### Figure
https://matplotlib.org/stable/api/figure_api.html?highlight=figure#module-matplotlib.figure

#### Subplot
https://matplotlib.org/stable/api/figure_api.html?highlight=figure#matplotlib.figure.SubplotParams

```python
figura = plt.figure(figsize=(20, 5))
figura.suptitle('Título')

#131 - 1 linha, 3 colunas, figura 1
#132 - 1 linha, 3 colunas, figura 2
figura.add_subplot(131)
plt.plot(x, y, label='dado azul')
plt.plot(y, x, label='dado vermelho', color='r')
plt.title('Gráfico 1')
plt.legend()

figura.add_subplot(132)
plt.bar(x, y, label='dado azul', color='y')
plt.title('Gráfico 2')
plt.legend()


figura.add_subplot(133)
plt.scatter(x, y, label='dado preto', color='k')
plt.title('Gráfico 3')
plt.legend()


plt.show(figura)
```
